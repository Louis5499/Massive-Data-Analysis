```python
from pyspark import SparkConf, SparkContext
```

```python
import math
def mapper1(line):
    linelist = line.splitlines()
    total_length = int(math.sqrt((len(line) + 1)/2))
    maplist = []
    for wordlist in linelist:
        tidied = wordlist.split(',')
        if tidied[0] == 'M':
            for idx in range(total_length):
                maplist.append(((int(tidied[1]), idx, int(tidied[2])), int(tidied[3])))
        else:
            for idx in range(total_length):
                maplist.append(((idx, int(tidied[2]), int(tidied[1])), int(tidied[3])))
    return maplist
            
def mapper2(lines):
    maplist = []
    tempKey = ''
    for resultVal in lines:
        if not tempKey:
            tempKey = resultVal
        else:
            maplist.append(((tempKey[0], tempKey[1]), resultVal))
            tempKey = ''
    return maplist;
```

```python
def reduce1(x, y):
    return x*y

def reduce2(x, y):
    return x+y
```

```python
conf = SparkConf().setMaster("local").setAppName("wordcount")
sc = SparkContext.getOrCreate(conf=conf)
lines = sc.textFile("./500input.txt").flatMap(mapper1)
# print(lines.collect())
lines = lines.reduceByKey(reduce1)
# print(lines.collect())
lines = lines.flatMap(mapper2)
# print(lines.collect())
lines = lines.reduceByKey(reduce2)
print(lines.collect())
sc.stop()
```
